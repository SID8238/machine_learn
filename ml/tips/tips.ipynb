```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
x
y = x **3
y
```

```python
plt.plot(x,y)
plt.title('Hello')
plt.xlabel(" x axis ")
plt.ylabel(" y axis ")
```

```python
plt.subplot(2,2,1)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
```

```python
fig = plt.figure(figsize=(12,8), dpi = 100)
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2 = fig.add_axes([0.4,0.4,0.3,0.5])
axis2.plot(x,y)
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("basicplot.png")
```
